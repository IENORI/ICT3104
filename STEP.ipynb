```python
%cd /content/drive/MyDrive/STEP/datasets/demo/frames

# Importing all necessary libraries
import cv2
import os

# Read the video from specified path
cam = cv2.VideoCapture("/content/drive/MyDrive/STEP/datasets/demo/"+video_name_wth_ext)

try:
	
	# creating a folder named data
	if not os.path.exists(video_name_wthout_ext):
		os.makedirs(video_name_wthout_ext)

# if not created then raise error
except OSError:
	print ('Error: Creating directory of data')

# frame
currentframe = 0

while(True):
	
	# reading from frame
	ret,frame = cam.read()

	if ret:
		# if video is still left continue creating images
		name = '/content/drive/MyDrive/STEP/datasets/demo/frames/'+video_name_wthout_ext+'/' + 'frame'+str(f'{currentframe:04}') + '.jpg'
		print ('Creating...' + name)

		# writing the extracted images
		cv2.imwrite(name, frame)

		# increasing counter so that it will
		# show how many frames are created
		currentframe += 1
	else:
		break

# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```
